## Nivell avançat

Desgraciadamente, los datos oficiales no están publicados para ser analizados... Vamos a trabajos sobre el documento: "Gasto de los turistas extranjeros por periodo y Comunidad Autónoma de destino principal" Fuente IBESTAT.

https://intranet.caib.es/ibestat-jaxi/tabla.do?px=68904221-f893-456c-8046-261150ba9a2e&pag=1&nodeId=%20f58f0937-c64f-469d-bad5-99f29bbb59ce%20&pxName=I208004_n100.px


```python
import pandas as pd
import numpy as np
df = pd.read_csv("IBESTAT_Gasto de los turistas extranjeros por periodo y Comunidad Autónoma de destino principal.csv",sep=";",header=1) #Malament: noms de fitxers amb espais i accents.
df.head(13)
```

```python
# Estructuració inadecuada de dades:
# La primera columna contiene información que debería ser extendida en dos columnas diferentes: fechas y comunidad.
# ¿Como podemos resolverlo?
dates = df[df.columns[0]].apply(lambda x: x if x[0].isdigit() else np.nan)
print(dates[:14])
```

```python
dates.fillna(method="ffill",inplace=True)
print(dates[:14])
```

```python
dates_ok = pd.to_datetime(dates.str.replace('M', '-', regex = True), format = '%Y-%m')
print(dates_ok[:14])
```

```python
df.index = pd.to_datetime(dates_ok)
df.head(13)
```

```python
np.unique(df[df.columns[0]]) ## MALAMENT: noms amb espais ("Andalucia" o "      Andalucia"?) 
```

```python
# Puedes crear una nueva columna con la información correspondiente de las Comunidades, ignora la no comunidad '   TOTAL'. !!!  Detalles... 
comunidades = df[df.columns[0]].apply(lambda x: x if (not x[0].isdigit()) and x != '   TOTAL' else np.nan)
comunidades.head(13)
```

```python
# crea una nueva columna con las comunidades

df["Comunidades"] = comunidades
df.head(13)
```

```python
# borra columnas: "Unnamed: 0", "Unnamed: 13"
df.drop(['Unnamed: 0',"Unnamed: 13"], axis=1,inplace=True) #
```

```python
# borrar las Filas innecesarias (las que no tienen comunidad: por Total o por fecha)
df = df.dropna()
```
